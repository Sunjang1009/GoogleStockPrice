```python
import re
from re import match, sub, findall
import matplotlib.pyplot as plt
from numpy.random import randn
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
from matplotlib import font_manager
import matplotlib


```

```python
#한글 글꼴 설정하였음.. matplotlib.rc 로 결정하는 모양 ㅋㅋ

font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
print(font_name)
matplotlib.rc('font',family='font_name')

s = Series(randn(10).cumsum(),index=np.arange(0,100,10))
print(s)
```

```python
s.plot()
plt.show()
```

```python
s.plot(label='test',
       style='ko--')

plt.show()
```

```python
s.plot(label='test',
       style='ko--',
       xticks=range(0,100,20),
       yticks=range(-5,5,2))

plt.show()

```

```python
## now using DataFrame..

df = DataFrame(np.random.rand(10,4), columns=list('ABCD'),index=np.arange(0,100,10))
print(df)

```
